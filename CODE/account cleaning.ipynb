```python
import pandas as pd
import numpy as np
import re
i=0
for df in pd.read_csv(r'S:\sonata\asssignment1\KOTAK\Account.csv',on_bad_lines='skip',chunksize=50000,sep='|',low_memory=False):
    df.dropna(how='all',axis=1,inplace=True)
    df.drop_duplicates(inplace=True)
    df=df.infer_objects()
    df.columns=df.columns.str.strip()
    df[dates]=df[dates].apply(pd.to_datetime,dayfirst=True,errors='coerce')
    ndf=df[df['LOS-APP-ID'].str.startswith('SFAC')]
    cols=['CREDT-RPT-ID','LOS-APP-ID','DISBURSED-DT','INSTALLMENT-AMT','MATCH-TYPE','WRITE-OFF-AMT','ACCT-TYPE','SELF-INDICATOR','ACCOUNT-STATUS','CREDIT-GRANTOR','OVERDUE-AMT','WRITE-OFF-DATE','LAST-PAYMENT-DATE']
    ndf=ndf[cols]
    if i==0:
        ndf.to_csv(r'S:\sonata\asssignment1\KOTAK\Final Files\PSF\ACCOUNT-PSF.csv',index=False)
        i+=1
    elif i!=0:
        ndf.to_csv(r'S:\sonata\asssignment1\KOTAK\Final Files\PSF\ACCOUNT-PSF.csv',header=False,index=False,mode='a')
        i+=1
print("done"i)
```
